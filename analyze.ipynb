### Imports

```python
import csv
from math import pi, sqrt
import matplotlib.pyplot as plt
import numpy as np
```

### Add datas

```python
t, y, vy, v, ad = range(5)

def Fd(u, r):
    Cd = 0.47
    airDensity = 1.293 #kg/m3
    return Cd * airDensity * u ** 2 * pi * r ** 2 

def Ad(u, r, m) : return Fd(u, r) / m

m = [0, 2, 2, 2, 2, 2]

a = [0.318] * 6
```

### Draw plots

```python
def draw(rows, i):
    ts = []
    ys = []
    vs = []
    ads = []
    for a, b, c, d, e in rows:
        ts.append(a)
        ys.append(b)
        vs.append(c)
        ads.append(e)
    plt.plot(ts, ys, "red")
    plt.show()
    plt.plot(ts, vs, "blue")
    plt.show()
    plt.plot(ts, ads, "green")
    plt.show()
```

```python
def useHeight(rows):
    maxHeights = []
    minHeights = []
    maxTime = []
    minTime = []
        
    for i in range(1, len(rows) - 1):
        if (rows[i][y] >= rows[i - 1][y] and rows[i][y] >= rows[i + 1][y]):
            maxHeights.append(rows[i][y])
            maxTime.append(rows[i][t])
        elif (rows[i][y] < rows[i - 1][y] and rows[i][y] < rows[i + 1][y]):
            minHeights.append(rows[i][y])
            minTime.append(rows[i][t])
    
    print(maxTime)
    print(minTime)
    
    maxHeights = list(set(maxHeights))
    minHeights = list(set(minHeights))
    
    heights = []
    if len(maxHeights) == len(minHeights):
        for i in range(1, len(maxHeights)):
            h1 = maxHeights[i - 1] - minHeights[i]
            h2 = maxHeights[i] - minHeights[i]
            heights.append((h1, h2))
    else:
        print("Not an available data!")
        return
    
    print("Max heights: ")
    print(maxHeights)
    print("Min heights: ")
    print(minHeights)
    
    print("Actual heights: ")
    for h in heights: print((round(h[0], 4), round(h[1], 4)))
    print("The result(not concering the Fd, and discard the result that isn't reasonable): ")
    results = []
    for h in heights: 
        result = sqrt(h[1] / h[0])
        if result < 1:
            results.append(result)
    print()
    std, avg = np.std(results), np.average(results)
    print("Standard:", std)
                
    print("Results: ")
    for r in results: print(r, end = " ")
    print()
```

```python
def analyze(file):
    with open(f"{file}.csv", 'r', encoding="utf-8") as f:
        #Read and change the format of the datas
        rows = csv.reader(f)
        rows = list(rows)[6:-2]
        for i in range(len(rows)):
            for j in range(len(rows[i])):
                rows[i][j] = float(rows[i][j])
                
        for i in range(len(rows)):
            rows[i].append(Ad(rows[i][vy], 0.318, 2))
                
        print(f"File: {file}.csv")
        draw(rows, file)
        
        #useVelocity(rows)
        
        useHeight(rows)
        
        print("-------------------------------------------")

for i in range(1, 7):
    analyze(i)
```
